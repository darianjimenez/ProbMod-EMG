```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Get all CSV files in the emg_data_4_sensor directory
data_dir = Path('emg_data_4_sensor')
csv_files = sorted(data_dir.glob('four_rep*.csv'))

print(f"Found {len(csv_files)} CSV files")
print(f"CSV files path: {data_dir.absolute()}\n")

# Select a file to plot (using the first one as example)
if csv_files:
    selected_file = csv_files[0]
    print(f"Plotting: {selected_file}")
    
    df = pd.read_csv(selected_file)
    
    plt.figure(figsize=(12, 6))
    plt.plot(df['timestamp'], df['emg_value_sensor1'], label='Sensor 1', linewidth=1.5)
    plt.plot(df['timestamp'], df['emg_value_sensor2'], label='Sensor 2', linewidth=1.5)
    plt.plot(df['timestamp'], df['emg_value_sensor3'], label='Sensor 3', linewidth=1.5)
    plt.plot(df['timestamp'], df['emg_value_sensor4'], label='Sensor 4', linewidth=1.5)
    
    plt.xlabel('Timestamp')
    plt.ylabel('EMG Value')
    plt.title(f"EMG Data from {selected_file.name}")
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("No CSV files found in the emg_data_4_sensor directory")
```
